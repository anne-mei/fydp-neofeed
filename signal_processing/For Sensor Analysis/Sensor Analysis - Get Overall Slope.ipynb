```python
%matplotlib nbagg
#Import required libraries
from serial import Serial
import csv
import time
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from scipy import stats
import pandas as pd
```

```python

def get_slope(x,y):
    slope = plot_slope(x,y)
    density = 1 #in g/mL
    time_conversion = 60
    flow_rate = np.multiply(slope,1/density*time_conversion)
    return flow_rate
def plot_slope(x,y):
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    def myfunc(x):
        return slope * x + intercept
    mymodel = list(map(myfunc, x))
    fig,ax = plt.subplots(1, 1)
    ax.plot(x,y)
    ax.plot(x, mymodel)
    plt.show()
    return slope
# def get_flowrate(mass_rate):
#     density = 1 #in g/mL
#     time_conversion = 60
#     flow_rate = mass_rate/density*time_conversion
#     return flow_rate
```
