```python
def plot_slope(x,y,ax,is_final):
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    def myfunc(x):
        return slope * x + intercept
    mymodel = list(map(myfunc, x))
    #ax.plot(x,y)
    ax.plot(x, mymodel)
    density_milk =1 
    sec_to_min = 60
    if is_final is False:
        slope = slope*sec_to_min*density_milk
    return slope

# def plot_slope2(x,y,ax,is_final,is_plot):
def plot_slope2(x,y,is_final):
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    def myfunc(x):
        return slope * x + intercept
    mymodel = list(map(myfunc, x))
    #ax.plot(x,y)
#     if is_plot:
#         ax.plot(x, mymodel,label = 'Experimental Results - Lin Reg')
    density_milk =1 
    sec_to_min = 60
    if is_final is False:
        slope = slope*sec_to_min*density_milk
    return slope, mymodel
```
