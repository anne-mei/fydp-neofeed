```python
%matplotlib nbagg
#Import required libraries
from serial import Serial
import csv
import time
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from drawnow import drawnow, figure
from scipy import stats
import scipy.signal as signal
import pandas as pd
```

```python
def plot_slope(x,y,ax):
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    def myfunc(x):
        return slope * x + intercept
    mymodel = list(map(myfunc, x))
    ax.plot(x,y)
    ax.plot(x, mymodel)
    return slope
```

# 89 cm

```python
#Read real data
csv_name = r'\Height Calibration Trials - 89.csv'
csv_path =r'C:\Users\a_hop\repos\fydp-neofeed\sensor_data\2022-02-22 Height Calibration Straight Tube'+csv_name
csv_data = pd.read_csv(csv_path)

time = csv_data['Time']

#Initialize plot
fig, (ax1, ax2,ax3) = plt.subplots(3)

trial_1_slope = plot_slope(time[csv_data['Trial 1'].notna()],csv_data['Trial 1'].dropna(),ax1)
trial_2_slope = plot_slope(time[csv_data['Trial 2'].notna()],csv_data['Trial 2'].dropna(),ax2)
trial_3_slope = plot_slope(time[csv_data['Trial 3'].notna()],csv_data['Trial 3'].dropna(),ax3)
#print(trial_1_slope)
avg_slope_89 = (1/2)*(trial_1_slope +trial_2_slope+trial_3_slope)

print("\nTrial 1 Slope: "+str(trial_1_slope))
print("\nTrial 2 Slope: "+ str(trial_2_slope))
print("\nTrial 3 Slope: "+ str(trial_3_slope))
print("\nAvg slope: "+ str(avg_slope_89))
```
