```python
%matplotlib nbagg
#Import required libraries
from serial import Serial
import csv
import time
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from drawnow import drawnow, figure
from scipy import stats
import scipy.signal as signal
import pandas as pd
```

```python
def plot_slope(x,y,ax):
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    def myfunc(x):
        return slope * x + intercept
    mymodel = list(map(myfunc, x))
    ax.plot(x,y)
    ax.plot(x, mymodel)
    return slope
```

# 68 cm (Highest Height)

```python
#Read real data
csv_name = r'\Calibration Testing Results - 68 cm.csv'
csv_path =r'C:\Users\a_hop\repos\fydp-neofeed\sensor_data\2022-02-05 Height Calibration'+csv_name
csv_data = pd.read_csv(csv_path)

time = csv_data['Time']

#Initialize plot
fig, (ax1, ax2,ax3) = plt.subplots(3)

trial_1_slope = plot_slope(time,csv_data['Trial 1'],ax1)
trial_2_slope = plot_slope(time,csv_data['Trial 2'],ax2)
trial_3_slope = plot_slope(time,csv_data['Trial 3'],ax3)
print(trial_1_slope)
avg_slope_68 = (1/2)*(trial_1_slope +trial_2_slope)

print("\nTrial 1 Slope: "+str(trial_1_slope))
print("\nTrial 2 Slope: "+ str(trial_2_slope))
print("\nTrial 3 Slope: "+ str(trial_3_slope))
print("\nAvg slope: "+ str(avg_slope_68))
```

# Final Calibration Slope

```python
avg_slopes = [avg_slope_68,avg_slope_58,avg_slope_48, avg_slope_38, avg_slope_28]
density_milk =1 
sec_to_min = 60
avg_flowrate = [x*sec_to_min*density_milk for x in avg_slopes]
print(avg_flowrate)
heights = [0.68,0.58,0.48,0.38,0.28]

fig3, ax1 = plt.subplots(1,1)
ax1.plot(heights,avg_flowrate)
ax1.set_xlabel('Height (cm)')
ax1.set_ylabel('Flow Rate (mL/min)')
regres_slope = plot_slope(heights,avg_flowrate,ax1)
print('Calibration slope: '+ str(regres_slope))
#print('Calibration intercept: '+str(regres_intercept))
```
