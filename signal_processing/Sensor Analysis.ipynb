```python
%matplotlib nbagg
#Import required libraries
from serial import Serial
import csv
import time
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
from scipy import stats
```

```python

def get_slope(csv_path,i1,i2):
    file = open(csv_path)
    csvreader = csv.reader(file)
    y = np.asarray(list(csvreader))
    y = y.flatten().astype(np.float)
    x = range(i1, i2, 1)
    y_new = y[i1:i2]
    slope = plot_slope(x,y_new)
    return slope
def plot_slope(x,y):
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    def myfunc(x):
        return slope * x + intercept
    mymodel = list(map(myfunc, x))
    fig,ax = plt.subplots(1, 1)
    ax.plot(x,y)
    ax.plot(x, mymodel)
    plt.show()
    return slope
```
